```python
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns;sns.set()
```

```python
rng = np.random.RandomState(42)
X   = np.dot(rng.rand(2,2), rng.randn(2,200)).T
plt.scatter(X[:,0], X[:,1], s=50, ec='k', fc='yellow', lw=1.2)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=0)
pca.fit(X)
```

```python
pca.components_, pca.explained_variance_
```

```python
pca.mean_, pca.explained_variance_ratio_
```

```python
def plot_vectors(v0, v1, ax=None):
  ax = ax or plt.gca()
  arrowprops = dict(
      arrowstyle='->',
      color='blue', lw=2.0,
      shrinkA=0, shrinkB=0
  )

  ax.annotate('', v1, v0, arrowprops=arrowprops)
```

```python
plt.scatter(X[:,0], X[:,1], s=50, ec='k', fc='yellow', lw=1.2)

for vec, len in zip(pca.components_, pca.explained_variance_):
  v = 3*vec*np.sqrt(len)
  plot_vectors(pca.mean_, pca.mean_+v)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1, random_state=0)
pca.fit(X)
```

```python
Xtrf = pca.transform(X)
Xinv = pca.inverse_transform(Xtrf)
```

```python
Xtrf.shape, Xinv.shape
```

```python
plt.scatter(X[:,0], X[:,1], s=50, ec='k', fc='yellow', lw=1.2, alpha=0.3)
plt.scatter(Xinv[:,0], Xinv[:,1], s=50, ec='k', fc='blue', lw=1.2)
```
